```python
# Import Dependencies
import pandas as pd
```

```python
# Reference the file where the CSV is located
df = pd.read_csv('crime_incident_data2017.csv')
df.head()
# Import the data into a Pandas DataFrame
```

```python
# look for missing values
df.count()
```

```python
df.isna().sum()
```

```python
# drop null rows
df = df.dropna(how='any')
```

```python
# verify counts
df.count()
```

```python
# Check to see if there are any values with mispelled or similar values in "Offense Type"
df['Offense Type'].value_counts()
```

```python
# Combine similar offenses
df['Offense Type'] = df['Offense Type'].replace({'Theft From Motor Vehicle': 'Motor Vehicle Theft', 
                                                 'Motor Vehicle Theft' : 'Motor Vehicle Theft',
                                                'Theft of Motor Vehicle Parts or Accessories':'Motor Vehicle Theft'
                                                })
```

```python
df['Offense Type'].value_counts()
```

```python
# Combine similar offenses
df['Offense Type'] = df['Offense Type'].replace({'Vandalism': 'Thefts and Burgalary', 
                                                 'Burglary' : 'Thefts and Burgalary',
                                                'Shoplifting':'Thefts and Burgalary',
                                                 'Theft From Building':'Thefts and Burgalary',
                                                 'Robbery':'Thefts and Burgalary',
                                                 'Pocket-Picking':'Thefts and Burgalary',
                                                 'Embezzlement':'Thefts and Burgalary',
                                                 'Stolen Property Offenses': 'Thefts and Burgalary'
                                
                                                })
```

```python
df['Offense Type'].value_counts()
```
